# Simulating real time streaming
### Reading data from csv file
### Append the timestamp that will serve as the event time for windowing
### Publish the event to pubsub topic
### Stream rate is hardcoded to 1 message per second

```python
from google.cloud import pubsub_v1
from time import sleep,time
publisher = pubsub_v1.PublisherClient()
topic_path = publisher.topic_path("your-project", "topic_name")
```

```python
with open(r"C:\projects\streaming_pipeline\stock_market_data.csv",'r') as file_pointer:
    header = file_pointer.readline()
    for line in file_pointer:
        data = line[:-1]+","+str(int(time()))+"\n"
        print("publishing {0} to {1}".format(data,topic_path))
        publisher.publish(topic_path,data.encode("utf8"))
        sleep(1)
        
```
